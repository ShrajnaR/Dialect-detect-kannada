```python
from pydub import AudioSegment
import librosa
import librosa.display
import matplotlib.pyplot as plt

def convert_mp3_to_wav(mp3_file, wav_file):
    audio = AudioSegment.from_mp3(mp3_file)
    audio.export(wav_file, format="wav")

def extract_mfcc(wav_file):
    audio, _ = librosa.load(wav_file, sr=None)
    mfccs = librosa.feature.mfcc(y=audio, sr=44100, n_mfcc=13)
    return mfccs

def plot_mfcc(mfccs):
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfccs, x_axis='time')
    plt.colorbar()
    plt.title('MFCCs')
    plt.show()

# Example usage
preprocessed_audio_file = "C:/Users/dell/Downloads/dakshinakannada500.mp3"
temp_wav_file = "C:/Users/dell/Documents/fdak_audio.wav"

# Convert MP3 to WAV
convert_mp3_to_wav(preprocessed_audio_file, temp_wav_file)

# Extract and plot MFCCs
mfccs = extract_mfcc(temp_wav_file)
plot_mfcc(mfccs)
```

```python
from pydub import AudioSegment
import librosa
import librosa.display
import matplotlib.pyplot as plt

def convert_mp3_to_wav(mp3_file, wav_file):
    audio = AudioSegment.from_mp3(mp3_file)
    audio.export(wav_file, format="wav")

def extract_mfcc(wav_file, target_sr=16000, n_mfcc=13, hop_length=512):
    audio, sr = librosa.load(wav_file, sr=target_sr)
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc, hop_length=hop_length)
    return mfccs

def plot_mfcc(mfccs):
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfccs, x_axis='time')
    plt.colorbar()
    plt.title('MFCCs')
    plt.show()

# Example usage
preprocessed_audio_file = "C:/Users/dell/Documents/pdak_audio.mp3"
temp_wav_file = "C:/Users/dell/Documents/ffdak_audio.wav"

# Convert MP3 to WAV
convert_mp3_to_wav(preprocessed_audio_file, temp_wav_file)

# Extract and plot MFCCs
mfccs = extract_mfcc(temp_wav_file)
plot_mfcc(mfccs)
```
